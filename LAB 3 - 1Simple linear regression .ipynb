```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the dataset
df=pd.read_csv(r'/home/suhasext/Downloads/ML Suhas/placement.csv')
df
```

```python
# Scatter plot of CGPA vs. Package
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python

# Features (CGPA) and Target (Package)
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
# Linear Regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# Predictions on the test data
y_predict = lr.predict(x_test)
# Print predicted values
print(y_predict)
```

```python
# Scatter plot and regression line
plt.scatter(df['cgpa'], df['package'])
plt.plot(x_train, lr.predict(x_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python


# Slope (m) and Intercept (b) of the regression line
m = lr.coef_
print("Slope, m = ", m)
b = lr.intercept_
print("Y-Intercept, c = ", b)
```

```python
# Predicted values using the regression equation
print("Prediction for CGPA 8.58:", m * 8.58 + b)
print("Prediction for CGPA 6.86:", m * 6.86 + b)
```

```python

# Calculate Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_predict)
print("Mean Squared Error is:", MSE)

# Calculate Root Mean Squared Error (RMSE)
RMSE = np.sqrt(MSE)
print("Root Mean Squared Error is:", RMSE)

# Calculate Mean Absolute Error (MAE)
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, y_predict)
print("Mean Absolute Error is:", MAE)
```
